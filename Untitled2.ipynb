```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x=np.array([1,2,3,4,5]).reshape((-1,1))
y=np.array([3,6,13,28,55])

```

```python
linear_model = LinearRegression()
linear_model.fit(x, y)
y_pred = linear_model.predict(x)
```

```python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
poly_model=LinearRegression()
poly_model.fit(x_poly,y)
y_poly_pred=poly_model.predict(x_poly)
```

```python
plt.scatter(x,y,color='red',label='Actual')
plt.plot(x,y_pred,label='Linear Fit',color='blue')
plt.plot(x,y_poly_pred,label='Polnominal Fit(deg=2)',color='green')
plt.title("Linear vs Polynominal Fit (Dummy Data)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Real Dataset:Salary prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Salary_Data.csv")
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
x=df[["Years of Experience"]]
y=df["Salary"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lin_model=LinearRegression()
lin_model.fit(x_train,y_train)
y_pred=lin_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

mse_lin=mean_squared_error(y_test,y_pred)
rmse_lin=np.sqrt(mse_lin)
r2_lin=r2_score(y_test,y_pred)
print("Root Mean Squared Error (Linear Model):",rmse_lin)
print("Mean Squared Error (Linear Model):",mse_lin)
print("R-squared (Linear Model):",r2_lin)
```

```python
poly2=PolynomialFeatures(degree=2)
x_train_poly2=poly2.fit_transform(x_train)
x_test_poly2=poly2.transform(x_test)
poly_model2=LinearRegression()
poly_model2.fit(x_train_poly2,y_train)
y_pred_poly2=poly_model2.predict(x_test_poly2)
```

```python
mse_poly2=mean_squared_error(y_test,y_pred_poly2)
rmse_poly2=np.sqrt(mse_poly2)
r2_poly2=r2_score(y_test,y_pred_poly2)
```

```python
poly4=PolynomialFeatures(degree=4)
x_train_poly4=poly4.fit_transform(x_train)
x_test_poly4=poly4.transform(x_test)

poly4_model=LinearRegression()
poly4_model.fit(x_train_poly4,y_train)
y_pred_poly4=poly4_model.predict(x_test_poly4)
```

```python
mse_poly4=mean_squared_error(y_test,y_pred_poly4)
rmse_poly4=np.sqrt(mse_poly4)
r2_poly4=r2_score(y_test,y_pred_poly4)
print("\nPolynominal Regression (Degree 2):")
print("Root Mean Squared Error (Linear Model):",rmse_lin)
print("Mean Squared Error (Linear Model):",mse_lin)
print("R-squared (Linear Model):",r2_lin)
print("\nPolynominal Regression (Degree 4):")
print("Root Mean Squared Error (Polynomial Model, Degree 2):",rmse_poly2)
print("Mean Squared Error (Polynomial Model, Degree 2):",mse_poly2)
print("R-squared (Polynomial Model, Degree 2):",r2_poly2)
```
